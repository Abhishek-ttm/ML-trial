```python
from sklearn import tree
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_excel('data/Diabetes_new.xlsx')
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['Patient number', 'Unnamed: 16', 'Unnamed: 17'])
df.isna().sum()
```

```python
X = df.drop(columns=['Diabetes'])
Y = df['Diabetes']
```

```python
X.info()
```

```python
X['Gender'] = X['Gender'].replace(['male','female'], [0,1])
```

```python
X.info()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
DTclf = tree.DecisionTreeClassifier()
DTclf = DTclf.fit(X_train, Y_train)
```

```python
predictions = DTclf.predict(X_test)
predictions
```

```python
DTclf.predict_proba(X_test)
```

```python
text_representation = tree.export_text(DTclf)
print(text_representation)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(20,20))
__ = tree.plot_tree(DTclf, feature_names=X.columns, class_names=Y, filled = True)
```
